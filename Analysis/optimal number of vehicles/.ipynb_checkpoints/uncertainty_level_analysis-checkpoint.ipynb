## Total Cost / Forecast Error analysis

```python
# Add column indicating the corresponding forecasting error
def f_err(dataf):
    if(dataf['Instance'][-5:] == '2.ser'):
        return 0.
    elif(dataf['Instance'][-7:] == '1.0.ser'):
        return 1.
    elif(dataf['Instance'][-7:] == '.75.ser'):
        return 0.75
    elif(dataf['Instance'][-7:] == '0.5.ser'):
        return 0.5
    elif(dataf['Instance'][-7:] == '.25.ser'):
        return 0.25
    
forecast['ForecastError'] = forecast.apply(f_err, axis = 1)
```

# Regression: forecasting error multiplier as predictive variable of the total cost

## Linear Regression

```python
from sklearn import linear_model
# Linear regression Cost~ForecastError
def linear_regression(data,cost):
    lr = linear_model.LinearRegression()
    reg = lr.fit(data['ForecastError'].values.reshape(-1,1), data[cost])
    plt.scatter(forecast_error*num_instances, data[cost].values.ravel(), color='black')
    plt.plot(forecast_error*num_instances, reg.predict(data['ForecastError'].values.reshape(-1,1)), color='blue',linewidth=1)
    # R^2
    print("R^2 value: "+str(reg.score(data['ForecastError'].values.reshape(-1,1), data[cost])))
    print("RSS"+str(reg.residues_))
```

```python
linear_regression(forecast,'TotalCost')
```

Add dummy variables to overcome different ranges of cost problem:

```python
# Add boolean columns
feature_list = []
for i in range(1, num_instances+1):
    col_name = ("b"+str(i))
    feature_list.append(col_name)

# Fill columns b1 to b(n-1), where n is the number of instances, with 1s in the appropriate position. 
# The last column remains 0
booleans = pd.DataFrame(0, index=forecast.index, columns=feature_list)
for i in range(0, num_instances-1):
    for j in range(0,5):
        booleans.iloc[i*(5)+j][i] = 1

# Add the booleans values to the forecast data
forecast_bool = pd.concat([forecast, booleans],axis=1)
```

```python
forecast_bool
```

```python
# Add forecast error as predictor in addition to the dummy variables
feature_list.append('ForecastError')
```

```python
# Linear regression Cost~[ForecastError,b1,...,bn]
def linear_regression_bool(data, cost):
    lr = linear_model.LinearRegression()
    reg = lr.fit(data[feature_list].values, data[cost].values)
    # R^2 
    reg.score(data[feature_list].values, data[cost].values)
    print("R^2 value: "+str(reg.score(data[feature_list].values, data[cost].values)))
    residuals = data[cost] - reg.predict(data[feature_list].values)
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((data[cost]-np.mean(data[cost]))**2)
    r_squared = 1 - (ss_res / ss_tot)
    print('RSS: '+ str(ss_res)) 
```

```python
linear_regression_bool(forecast_bool, 'TotalCost')
```

```python
linear_regression_bool(forecast_bool, 'RoutingCost')
```

```python
linear_regression_bool(forecast_bool, 'OverflowCost')
```

### Exponential  Fit

We run a non-linear regression using an exponential function.

```python
from scipy.optimize import curve_fit
# Exponential function
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
```

```python
# Non-linear regression Cost~forecastError
def exp_fit(cost):
    # popt gives value for a, b and c. 
    popt, pcov = curve_fit(func, forecast['ForecastError'].values.ravel(), forecast[cost])
    # difference between actual cost and cost forecasted by function
    residuals = forecast[cost] - func(forecast['ForecastError'].values.ravel(), popt[0], popt[1], popt[2])
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((forecast[cost]-np.mean(forecast[cost]))**2)
    r_squared = 1 - (ss_res / ss_tot)
    print('RSS: '+ str(ss_res))
    plt.scatter(forecast_error, averages[cost]);
    plt.plot(np.arange(0, 1.01, 0.01), func(np.arange(0, 1.01, 0.01), popt[0], popt[1], popt[2]), color='blue',linewidth=1)
```

```python
exp_fit('TotalCost')
```

```python
exp_fit('RoutingCost')
```

```python
exp_fit('OverflowCost')
```

### Plot linear and non-linear regressions

```python
pylab.rcParams['figure.figsize'] = 5, 5
```

```python
def fit(cost):
    # Exponential fit
    popt, pcov = curve_fit(func, forecast['ForecastError'].values.ravel(), forecast[cost])
    residuals = forecast[cost] - func(forecast['ForecastError'].values.ravel(), popt[0], popt[1], popt[2])
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((forecast[cost]-np.mean(forecast[cost]))**2)
    r_squared = 1 - (ss_res / ss_tot)
    print('Exp RSS: '+ str(ss_res))
    plt.plot(np.arange(0, 1.01, 0.01), func(np.arange(0, 1.01, 0.01), popt[0], popt[1], popt[2]), color='#50c94c',linewidth=3, label='Exponential Fit')
    p = plt.boxplot([forecast0, forecast25, forecast50, forecast75, forecast1], positions=forecast_error, widths=[0.03]*5, manage_xticks=False,patch_artist=True)
    colors = ['#dce4f7']*5
    for patch, color in zip(p['boxes'], colors):
        patch.set_facecolor(color)
        patch.set_color(color)
    plt.xticks(forecast_error)
    
    # Linear fit
    lr = linear_model.LinearRegression()
    reg = lr.fit(forecast['ForecastError'].values.reshape(-1,1), forecast[cost])
    plt.plot(forecast_error*num_instances, reg.predict(forecast['ForecastError'].values.reshape(-1,1)), color='#0090bc',linewidth=1, label='Linear Fit')
    plt.xlabel("Forecasting error multipliers")
    plt.ylabel("Average cost (CHF)")
    plt.legend(loc=4)
    print('Linear RSS: '+str(reg.residues_))
    plt.savefig('fits.png',bbox_inches = 'tight')
```

```python
fit('TotalCost')
```

## Exponential fit with dummy variables

```python
# b is now the forecast error with the dummy variables
def func_exp(x, a, c, *b):
    return a * np.exp(np.hstack(b).dot(x)) + c
```

```python
# Cost~[ForecastError,b1,...,bn]
def exp_fit_bool(cost):
    popt, pcov = curve_fit(func_exp, forecast_bool[feature_list].T, forecast_bool[cost],p0=np.zeros(len(feature_list)+2))
    residuals = forecast_bool[cost] - func_exp(forecast_bool[feature_list].T, popt[0], popt[1], *(popt[2:len(popt)].ravel()))
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((forecast_bool[cost]-np.mean(forecast_bool[cost]))**2)
    r_squared = 1 - (ss_res / ss_tot)
    print('RSS: '+ str(ss_res))
```

```python
exp_fit_bool('TotalCost')
```
